```python
import random
import matplotlib.pyplot as plt

def linear_search_with_comparisons(sequence, value):
    comparisons = 0
    for item in sequence:
        comparisons += 1
        if item == value:
            return comparisons
    return comparisons  # Return total comparisons even if not found

def generate_sequence(length):
    return random.sample(range(1, length + 1),length)  # Ordered integers for simplicity

def plot_comparisons(x, y_best, y_avg, y_worst):
    plt.plot(x, y_best, marker='o', label='Best Case (Start)')
    plt.plot(x, y_avg, marker='o', label='Average Case (Middle)')
    plt.plot(x, y_worst, marker='o', label='Worst Case (Not Found)')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Number of Comparisons')
    plt.title('Linear Search: Comparisons vs Input Size')
    plt.legend()
    plt.grid(True)
    plt.show()

# Input sizes
input_sizes = [1000, 5000, 10000, 50000, 100000,200000]

# Store comparisons
comparison_counts = {
    'Best': [],
    'Average': [],
    'Worst': []
}

# Run experiment
for n in input_sizes:
    sequence = generate_sequence(n)

    # Best case: value at the beginning
    value_start = sequence[0]
    comparisons_start = linear_search_with_comparisons(sequence, value_start)
    comparison_counts['Best'].append(comparisons_start)

    # Average case: value in the middle
    value_middle = sequence[n // 2]
    comparisons_middle = linear_search_with_comparisons(sequence, value_middle)
    comparison_counts['Average'].append(comparisons_middle)

    # Worst case: value not in the list
    value_not_found = -1
    comparisons_worst = linear_search_with_comparisons(sequence, value_not_found)
    comparison_counts['Worst'].append(comparisons_worst)

# Display results in table
print("Input Size | Best Case | Average Case | Worst Case")
print("-----------|-----------|--------------|-----------")
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:>10} | {comparison_counts['Best'][i]:>9} | {comparison_counts['Average'][i]:>12} | {comparison_counts['Worst'][i]:>9}")

# Plotting
plot_comparisons(
    input_sizes,
    comparison_counts['Best'],
    comparison_counts['Average'],
    comparison_counts['Worst']
)
```
